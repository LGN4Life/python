```python
import requests
from bs4 import BeautifulSoup as soup

import re #regular expressions

import pdb

import csv

import os
```

```python
#get info from website
current_url = 'https://en.wikipedia.org/wiki/List_of_cities_and_towns_in_California'
result = requests.get(current_url)

src = result.content
#soup = BeautifulSoup(src, 'lxml')
city_soup = soup(src, 'html.parser')


output_file_name = 'C:\\Henry\\Hack4California\\PythonData\\CaliforniaCityData\\california_city_full_list.csv'
```

```python
#There is a table on the wiki page that has all the cities of California listed, with a link the the city's wiki page

#scrape table
city_data = city_soup.findAll('table',{'class': 'wikitable plainrowheaders sortable'})

#scrape city rows
city_rows = city_data[0].findAll('th',{'scope': 'row'})
```
